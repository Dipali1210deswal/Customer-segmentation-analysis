```python

columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in columns:
    plt.figure()
    g = sns.FacetGrid(df, hue='Gender', height=6, aspect=1.5)
    g.map(sns.kdeplot, i, fill=True).add_legend()
    g.set_axis_labels(i, 'Density')
    g.fig.suptitle(f'Distribution of {i}')
    g.fig.set_size_inches(8, 4)
    plt.show()
```

```python
palette = {'Male': 'lightblue', 'Female': 'lightcoral'}

# Create bar plots for each column
columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.barplot(data=df, x='Gender', y=i, palette=palette, ci=None)
    plt.title(f'Mean {i} by Gender')
    plt.show()
```

```python
age_bins = pd.cut(df['Age'], bins=[0, 20, 40, 60, 80], labels=['1-20', '21-40', '41-60', '61-80'])

# Count the number of occurrences in each bin
age_distribution = age_bins.value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=age_distribution.index, y=age_distribution.values, palette='coolwarm')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Distribution in Different Intervals')
plt.show()
```

```python
income_bins = pd.cut(df['Annual Income (k$)'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])

# Count the number of occurrences in each income bin
income_distribution = income_bins.value_counts().sort_index()

# Create a bar plot for annual income distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=income_distribution.index, y=income_distribution.values, palette='coolwarm')
plt.xlabel('Annual Income Group (k$)')
plt.ylabel('Count')
plt.title('Annual Income Distribution in Different Intervals')
plt.show()
```

```python
clustering1 = KMeans(n_clusters=3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
intertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    intertia_scores.append(kmeans.inertia_)
```

```python
intertia_scores
```

```python
plt.plot(range(1,11),intertia_scores)
```

```python
df.columns
```

```python
mean_values = df.groupby('Income Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
print(mean_values)
```

```python
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
df['Spending and Income Cluster'] =clustering2.labels_
df.head()
```

```python
intertia_scores2=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    intertia_scores2.append(kmeans2.inertia_)
plt.plot(range(1,11),intertia_scores2)
```

```python
centers =pd.DataFrame(clustering2.cluster_centers_)
centers.columns = ['x','y']
plt.figure(figsize=(8,5))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c='black',marker='*')
sns.scatterplot(data=df, x ='Annual Income (k$)',y='Spending Score (1-100)',hue='Spending and Income Cluster',palette='tab10')
plt.savefig('clustering_bivaraiate.png')
```

```python
#mulivariate clustering 
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
df.head()
```

```python
dff = pd.get_dummies(df,drop_first=True)
dff.head()
```

```python
dff.columns
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)','Gender_Male']]
dff.head()
```

```python
dff = scale.fit_transform(dff)
dff = pd.DataFrame(scale.fit_transform(dff))
dff.head()
```

```python
intertia_scores3=[]
for i in range(1,11):
    kmeans3=KMeans(n_clusters=i)
    kmeans3.fit(dff)
    intertia_scores3.append(kmeans3.inertia_)
plt.plot(range(1,11),intertia_scores3)
```
